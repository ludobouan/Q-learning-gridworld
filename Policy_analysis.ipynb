```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_eps_greedy_traces = pd.read_csv('data/log_eps_greedy_traces.csv', header=0)
df_original = pd.read_csv('data/log_original.csv', header=0)
df_greedy_traces = pd.read_csv('data/log_greedy_traces.csv', header=0)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=True)
fig.set_size_inches(12, 5)
ax1.set_title('Original (Greedy policy)'); ax2.set_title('Greedy with traces'); ax3.set_title('Epsilon-greedy with traces')
sns.regplot('episode', 'score', data=df_original, fit_reg=False, ax=ax1)
sns.regplot('episode', 'score', data=df_greedy_traces, fit_reg=False, ax=ax2)
sns.regplot('episode', 'score', data=df_eps_greedy_traces, fit_reg=False, ax=ax3)

# fig.savefig('data/Compare.png')
```
